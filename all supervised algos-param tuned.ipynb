```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Microsoft_Data.csv')
df.head()
```

```python
df.get_dtype_counts()
```

```python
#unoque value cols
cols_to_drop=[]
for col in df.columns:
    if len(df[col].value_counts())==1 or len(df[col].value_counts())==len(df):
        cols_to_drop.append(col)
cols_to_drop
```

```python
missing_df=pd.DataFrame({'col_name':df.columns,'count':df.isnull().sum(),'percentage':df.isnull().sum()/len(df)*100})
missing_df.sort_values(by='percentage',inplace=True,ascending=False)
missing_df
```

```python
cols_to_drop2=list(missing_df[missing_df['percentage']>45].index)
cols_to_drop2
```

```python
df.drop(columns=set(cols_to_drop+cols_to_drop2),inplace=True)
```

```python
from scipy.stats import chi2_contingency
table=[]
chi2=[]
pvalue=[]
no_corr=[]
for i in range(len(df.columns)):
    table.append(pd.crosstab(df.iloc[:,i],df.iloc[:,-1]))
    chi2.append(chi2_contingency(table[i]))
    pvalue.append(chi2[i][1])
    if pvalue[i]>0.05:
        no_corr.append(df.columns[i])
no_corr
```

```python
df.drop(columns=no_corr,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df['HasDetections']
df.drop(columns='HasDetections',inplace=True)

```

```python
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=.25,random_state=42)
```

```python
print('x_train shape:',x_train.shape)
print('x_test shape:',x_test.shape)
print('y_train shape:',y_train.shape)
print('y_test shape:',y_test.shape)
```

```python
#replae missing vaues
for col in df.columns:
    x_train[col]=x_train[col].fillna(x_train[col].mode()[0])
    x_test[col]=x_test[col].fillna(x_train[col].mode()[0])
```

```python
cat_cols=list(df.select_dtypes(include=[np.object]).columns)
num_cols=list(df.select_dtypes(include=[np.number]).columns)
```

```python
oe_x_train=pd.get_dummies(x_train)
oe_x_test=pd.get_dummies(x_test)
```

```python
final_x_train,final_x_test=oe_x_train.align(oe_x_test,axis=1,join='inner',fill_value=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import f1_score,classification_report,confusion_matrix,roc_auc_score,roc_curve,accuracy_score
```

```python
logistic=LogisticRegression(tol=0.001)
logistic.fit(final_x_train,y_train)
logistic_pre=logistic.predict(final_x_test)
print(classification_report(y_test,logistic_pre))
```

```python
logistic_prob=logistic.predict_proba(final_x_test)[::,1] # Probability of a record belonging to the positive class(class=1)

Lfpr,Ltpr,Lthreshold=roc_curve(y_test,logistic_prob) # the function roc_curve returns three values, the False Positive Rate,True Positive rate
                                                # threshold values it used to calculate the TPR and FPR
Lauc = roc_auc_score(y_test, logistic_prob)
print('the accuracy of logistic regression is: ',accuracy_score(y_test,logistic_pre))
print('the auc score of logisic regression is :',Lauc)
```
